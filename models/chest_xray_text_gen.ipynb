```python
import gpt_2_simple as gpt2
from datetime import datetime
```

```python
model_name = "124M"
gpt2.download_gpt2(model_name=model_name)
```

```python
sess = gpt2.start_tf_sess()
gpt2.finetune(sess,
              dataset_file,
              model_name=model_name,
              steps=200,  # Adjust the number of steps based on your dataset size and computing resources
              run_name='run1',
              print_every=10,
              sample_every=20  # Generate a sample every 20 steps
              )
```

```python
query = "The image shows signs of pneumonia with: Perimeter: 8052.698271155357 mm, Area: 531891.5 sqmm, Equivalent Diameter: 822.9369910925761 mm, Irregularity Index: 3.1147627314221884, Standard Deviation: 57.41007621497808. Give medicine recommendation and Prescription for it only."

generated_text = gpt2.generate(sess,
                              model_name=model_name,
                              length=20,
                              temperature=0.1,
                              top_p = 0.1,
                              prefix='Medicine Recommendations',
                              nsamples=1,
                              batch_size=1)
# print(generated_text)
```
